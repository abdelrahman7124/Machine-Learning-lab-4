## 2) Minimal Takeaways
- **Best clustering:** Exp 3 (K-Means + PCA, 15 PCs) — ARI 0.6594 in 0.13s.
- **Fastest:** Exp 1 (K-Means original) — 0.1s with near-best quality.
- **Dimensionality reduction helps:** PCA and AE improve clustering quality significantly.
- **GMM underperforms:** GMM generally yields lower ARI and NMI compared to K-Means.
- **AE benefits:** Autoencoder-based methods show promise but at higher computational cost.
